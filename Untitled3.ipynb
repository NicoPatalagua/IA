```
import random
from pyeasyga import pyeasyga

objetos = [(1, 4), (2, 4), (3, 3), (4, 2), (8, 7), (24, 10)]

peso = int(input("Digite el peso maximo de la mochila: "))

ga = pyeasyga.GeneticAlgorithm(
    objetos,
    population_size=200,
    generations=100,
    crossover_probability=0.8,
    mutation_probability=0.2,
    elitism=True,
    maximise_fitness=True
)


def fitness(individual, data):
    global peso
    pesot = peso
    pesol = 0
    beneficiol = 0
    for objeto in individual:
        pesol = pesol + (list(objeto))[0]
        beneficiol = beneficiol + (list(objeto))[1]
    puntuacion = beneficiol - abs(pesot - pesol)
    return puntuacion


def crossover(parent_1, parent_2):
    if len(parent_1) is not 0 and len(parent_2) is not 0:
        crossover_index = random.randrange(len(parent_1))
        child_1a = parent_1[:crossover_index]
        child_1b = [i for i in parent_2 if i not in child_1a]
        child_1 = child_1a + child_1b
        child_2a = parent_2[crossover_index:]
        child_2b = [i for i in parent_1 if i not in child_2a]
        child_2 = child_2a + child_2b
        return child_1, child_2
    else:
        return parent_1, parent_2


def mutate(individual):
    if len(individual) is not 0:
        iremocion = random.randrange(len(individual))
        individual.pop(iremocion)


def create_individual(data):
    individual = data[:]
    random.shuffle(individual)
    return individual


def selection(population):
    ruleta = []
    for dude in population:
        fitnes = int(dude.fitness)
        prob = int(100 + int(fitnes))
        ruleta = [dude for n in range(prob + 1)]
    return random.choice(ruleta)


ga.create_individual = create_individual
ga.crossover_function = crossover
ga.mutate_function = mutate
ga.selection_function = selection
ga.fitness_function = fitness
ga.create_first_generation()

for i in range(1, 100 + 1):
    ga.create_next_generation()

ga.run()
print("Mejor solucion: " + str(ga.best_individual()))
```
