```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
```

<h3>vamos a cargar el dataset</h3>

```python
##happy sad continuos
dataset=pd.read_csv("Dataset\HappySadContinuous.csv")
print(dataset.values[1662:1667,:])
#imp=SimpleImputer(missing_values='nan',strategy='mean')
#imp.fit(dataset)
dataset=dataset.fillna(0)
print(dataset.values[1662:1667,:])
```

```python
dataset.head()
```

```python
features=dataset.values[:,0:7]#vector de caracteristicas
labels=dataset.values[:,7]#vector de etiquetas/clases
```

<h3>Split Train/Test</h3>

```python
##aca dividimos nuestro dataset en entrenamiento y texto
#notese que no esta usando crossValidation que es el recomendado
x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=0.2,random_state=0)
```

<h3>Construccion del modelo</h3>

```python
###ahora entrenamos un modelo simple. Ya conocemos los arboles de desicion!
clf_entropy=DecisionTreeClassifier(criterion="entropy",random_state=100,max_depth=3,min_samples_leaf=5)
clf_entropy.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
import graphviz
%matplotlib inline
dot_data=tree.export_graphviz(clf_entropy,out_file=None)
graph=graphviz.Source(dot_data)
graph.render("happysad")
```

<h3>evaluacion</h3>

```python
### ahora vamos a evaluar que tan bueno es nuestro arbol, utilizando el set de pruebas
predictions=clf_entropy.predict(x_test)
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
print(accuracy_score(y_test,predictions))
```

<h3>Ahora con SVM</h3>

```python
from sklearn.svm import SVC
svm=SVC(C=1.0,kernel='linear',gamma='auto')
svm.fit(x_train,y_train)
predictions=svm.predict(x_test)
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
print(accuracy_score(y_test,predictions))
```
