```
import random
import matplotlib.pyplot as plt
```

Definimos el ingreso de la dimensión de la grilla:

```
n = int(input("Digite el tamaño de la stack: "))
Stack = [i for i in range(1, n + 1)]
```

Definimos la función fitness:

```
def fitness(indivudal):
    consecutivos = 0
    for i in range(len(indivudal)):
        if i > 0:
            if (indivudal[i] - 1) == (indivudal[i - 1]):
                consecutivos += 1
    return consecutivos
```

Definimos el algoritmo Bring to Top:

```

def BringToTop(individual):
    #print ("Inicio")
    mas=len(individual)
    if mas%2==0:
        mas=3
    else:
        mas=2

    for i in range(len(individual)-mas):
        max=i
        for j in range(len(individual)-i):
            if individual[max]<individual[j]:
                max=j
        aux=individual[max]
        individual.pop(max)
        individual.insert(0, aux)
        mitad=int((len(individual)-i)/2)
        for j in range(mitad):
            aux=individual[j]
            individual[j]=individual[len(individual)-i-j-1]
            individual[len(individual)-i-j-1]=aux
        print(individual)
    return individual
```

```
Definimos la creación del primer individuo:
```

```
def create_individual(data):
    individual = data[:]
    random.shuffle(individual)
    return individual
```

Creamos la pila inicial:

```
Stack = create_individual(Stack)
print(Stack)
mejorposible = (n - 1)
```

Imprimimos la pila ordenada:

```
Stack1 = BringToTop(Stack)
```

Realizamos los calculos correspondientes:

```
mejorobtenido = int(fitness(Stack))
prcentaje = float((100 * mejorobtenido) / mejorposible)
```
