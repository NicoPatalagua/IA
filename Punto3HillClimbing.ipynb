```
import random
from copy import deepcopy
import matplotlib.pyplot as plt
```

```
def fitness(indivudal):
    consecutivos = 0
    for i in range(len(indivudal)):
        if i > 0:
            if (indivudal[i] - 1) == (indivudal[i - 1]):
                consecutivos += 1
    return consecutivos
```

```
def mutate(individual):
    mi = random.randrange(len(individual))
    auxlist = individual[:mi]
    auxlist.reverse()
    for i in range(len(auxlist)):
        individual[i] = auxlist[i]
    return individual
```

```
def create_individual(data):
    individual = data[:]
    random.shuffle(individual)
    return individual
```

```
n = int(input("Digite el tama침o de la stack: "))
Stack = [i for i in range(1, n + 1)]
Stack = create_individual(Stack)
```

```
ejey = []
ejex = []

for i in range(1000):
    Stacki = deepcopy(Stack)
    Stacka = mutate(Stack)
    if fitness(Stacka) >= fitness(Stacki):
        Stack = deepcopy(Stacka)
    elif fitness(Stacka) < fitness(Stacki):
        Stack = deepcopy(Stacki)
    ejey.append(float(fitness(Stack)))
    ejex.append(float(format(i)))
```

```
mejorposible = (n - 1)
mejorobtenido = int(fitness(Stack))
prcentaje = float((100 * mejorobtenido) / mejorposible)
```

```
print("Porcentaje de acierto obtenido: " + str(prcentaje) + "%")
print("Fitness de la mejor solucion: " + str(mejorobtenido))
print("Mejor fitness posible: " + str(mejorposible))
print("Solucion: " + str(Stack) + "\n")
```

```
plt.plot(ejex, ejey, '-', linewidth=0.8, color='r', label='Fitness')
plt.title('Evoluci칩n costo/fitness')
plt.xlabel('Iteracion')
plt.ylabel('Fitness')
plt.legend()
plt.grid()
plt.show()
```
