###Función create_individual

Función de creación de individuos para la obtención de la población iniciail.

```
def create_individual(data):
    individual = data[:]
    longitud = len(individual)
    if (len(individual)) == 10:
        a = random.randrange(longitud)
        (random.choice(individual.pop(a)))
        random.shuffle(individual)
        (random.choice(individual.pop(a-1)))
        random.shuffle(individual)
        (random.choice(individual.pop(a-2)))
        individual.sort()
    elif (len(individual)) == 15:
        b = random.randrange(longitud)
        (random.choice(individual.pop(b)))
        random.shuffle(individual)
        (random.choice(individual.pop(b-1)))
        random.shuffle(individual)
        (random.choice(individual.pop(b-2)))
        random.shuffle(individual)
        (random.choice(individual.pop()))
        individual.sort()
    elif (len(individual))== 5:
        c = random.randrange(longitud)
        (random.choice(individual.pop(c)))
    return individual
```

##Función crossover

Con la cual realizamos diversos cruces entre las generaciones.

```
def crossover(parent_1, parent_2):
    crossover_index = random.randrange(1, len(parent_1))
    child_1a = parent_1[:crossover_index]
    child_2b = parent_2[crossover_index:]
    child_1 = child_1a + child_2b
    veces = 0
    for i in child_1:
        if child_1 == i:
            veces += 1
    child_1.sort()

    crossover_index1 = random.randrange(1, len(parent_2))
    child_2a = parent_2[:crossover_index1]
    child_1b = parent_1[crossover_index1:]
    child_2 = child_2a + child_1b
    veces1 = 0
    for i in child_2:
        if child_2 == i:
            veces1 +=1
    child_2.sort()

    return child_1,child_2
```

Con el objetivo de producir una diversidad en la población de estaciones de cada localidad.

```
def mutate(individual):
    long = (len(individual))
    a = 0
    if long == 10:
        for objeto in individual:
            a = (list(objeto))[0]
            if a == 1 or a == 3 or a == 7:
                mutate_index1a = random.randrange(1, len(individual))
                mutate_index2a = random.randrange(1, len(individual))
                mutate_num1a = random.randrange(500, 2000, 500)
                individual.insert(mutate_index1a, (mutate_index2a, mutate_num1a))
                individual.sort()
                individual.pop(mutate_index1a)
            elif a == 20 or a == 29:
                mutate_index1b = random.randrange(1, len(individual))
                mutate_index2b = random.randrange(16, 25, 1)
                mutate_num1b = random.randrange(500, 2000, 500)
                individual.insert(mutate_index1b, (mutate_index2b, mutate_num1b))
                individual.sort()
                individual.pop(mutate_index1b)
            elif a == 30 or a == 49:
                mutate_index1c = random.randrange(1, len(individual))
                mutate_index2c = random.randrange(41, 50, 1)
                mutate_num1c = random.randrange(500, 2000, 500)
                individual.insert(mutate_index1c, (mutate_index2c, mutate_num1c))
                individual.sort()
                individual.pop(mutate_index1c)
    elif long == 4:
        mutate_index3 = random.randrange(1, len(individual))
        mutate_index4 = random.randrange(11, 15, 1)
        mutate_num2 = random.randrange(500, 2000, 500)
        individual.insert(mutate_index3, (mutate_index4, mutate_num2))
        individual.sort()
        individual.pop(mutate_index3)
    elif long == 11:
        mutate_index5 = random.randrange(1, len(individual))
        mutate_index6 = random.randrange(26, 40, 1)
        mutate_num3 = random.randrange(500, 2000, 500)
        individual.insert(mutate_index5, (mutate_index6, mutate_num3))
        individual.sort()
        individual.pop(mutate_index5)
```
