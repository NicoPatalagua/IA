##Representación

La representación que se le da a un individuo de este ejercio debe bazarse en una matriz n × n, la cual sera una agrupación retangular de números con n filas y n columnas; La representación del ejemplo propuesto sería el siguiente:

##Fitness

En este problema se debe evaluar que losindividuos cumplen con todas las retricciones establecidas, es decir se debe garantizar que al ser sumadas las columnas, las filas y las diagonales de la matriz, deben arrojar el mismo valor. 
La función fitness de este algoritmo genético debería evaluar la función que 
especifica la dimensión de la grilla y la suma común.
